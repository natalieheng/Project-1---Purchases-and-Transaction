```python
#changing the data types
#1: date to dmy
transac_data['Date'] = pd.to_datetime(transac_data['Date'], format = '%d/%m/%y')
```

```python
#2: description of purchase and category of purchase as string
transac_data[['Description of Purchase', 'Category of Purchase']] = transac_data[['Description of Purchase', 'Category of Purchase']].astype('string')
```

# Data Analysis

```python
#getting rid of whitespace 
transac_data.columns = transac_data.columns.str.strip()
```

```python
#total spending for each month
transac_data['Month'] = transac_data['Date'].dt.strftime('%m')
```

```python
transac_data
```

```python
monthly_spending = transac_data.groupby(by = 'Month').sum()
```

```python
monthly_spending
```

```python
#money spent on each category each month? 
catspending_month = transac_data.groupby(by = ['Month', 'Category of Purchase']).sum()
```

```python
catspending_month
```

```python
#total spending (for the 3 month in total) for each category
cat_spendings = transac_data.groupby(by = 'Category of Purchase').sum()
```

```python
cat_spendings
```

```python
cat_spendings.dtypes
```

```python
#remove the extra spacing in the variable 'Price'
cat_spendings.columns = cat_spendings.columns.str.strip()
```

```python
cat_spendings.get('Price')
```

```python
cat_spendings['Price']
```

```python
cat_spendings.sort_values(by ='Price', axis = 0, ascending = False)
```
