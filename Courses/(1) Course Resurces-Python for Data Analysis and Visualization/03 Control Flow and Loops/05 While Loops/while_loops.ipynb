# While Loops

While Loops execute a block of code in a loop as long as the condition provided returns the boolean value True.

```python
# printing the statement and incrementing a by 1 for each loop
# The loop continues until a = 5

a=0
while a<5:
    print('a is less than 5')
    a+=1
```

```python
# incrementing a by 1 and then printing the statement
# The loop continues until a = 5

a=0
while a<5:
    a+=1
    print('a is less than 5')
```

```python
# incrementing a by 1 and then printing the statement stating the value of a
# The loop continues until a = 5

a=0
while a<5:
    a+=1
    print(f'the value of a is {a}')
```

Printing the statement before incrementing the value of a returns a different result where a is initially printed as 0

```python
# incrementing a by 1 and then printing the statement stating the value of a
# The loop continues until a = 5

a=0
while a<5:
    print(f'the value of a is {a}')
    a+=1 
```

```python
# This will create an infinite loop as the condition is always True
# To exit an infinite loop click on Kernel and then Interupt on the top Ribbon

while True:
    print('To infinity and beyond')
    break
```
