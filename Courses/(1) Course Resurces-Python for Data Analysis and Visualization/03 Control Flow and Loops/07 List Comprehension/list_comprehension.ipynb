# List Comprehension

```python
hw_text = 'hello world'
hw_list = []
```

```python
# You can append elements from a list object using a Loop and the append method
for character in hw_text:
    hw_list.append(character)
```

```python
hw_list
```

```python
# List Comprehension allows for simpler syntax
hw_list = [character for character in hw_text]
```

```python
hw_list
```

```python
# You can also perform operations and methods on the iterable element
hw_list = [character.upper() for character in hw_text]
```

```python
hw_list
```

```python
# You can also perform operations and methods on the iterable element
hw_list = [character+character for character in hw_text]
```

```python
hw_list
```

```python
hw_list =[]
```

```python
# This is the same as [character+character for character in hw_text]
for character in hw_text:
    hw_list.append(character+character)
```
