# if elif else

## if statement

The if-statement will run the block of code indented directly below if the condition returns True

```python
# The indented block of code will only run if the condition is True
# The second print function is not indented so it is not bound by the if-condition
if 4>3:
    print('the condition is true')
print('this statement will always print')
```

```python
x=60
```

```python
if x > 50:
    print(f'120% of the value of x is {x*1.2}')
```

```python
# The boolean True is case sensitive
if true:
    print('hello')
```

## if-else statement

The else condition will run as long as the preceding if condition is false

```python
x = 300
```

```python
if x < 200:
    print(' x is less than 200')
else:
    print('x is not less than 200')
```

## if-elif-else

You can add additional conditions by using the elif keyword in your if-elif-else statement.

The code gets evaluated from top to bottom and the block of code associated with the first condition that returns True is run.

```python
# Change the operators around in the below conditions to see which print function executes
if 5>100:
    print('The if statement is True')
elif 5>100:
    print('The elif statement is True')
elif 5>100:
    print('The second elif statement is True')
else:
    print('all conditions are False')
```

```python
if 5<100:
    print('The if statement is True')
if 5<100:
    print('The second if statement is True')
if 5<100:
    print('The third if statement is True')
```

```python
if x>=10:
    x = x*100
```

## nested statements

You can nest if, if-elif, and if-elif-else statements like in the below example

```python
x = 6
```

```python
if x<11:
    if x>=5:
        print('x is less than 11 but greter than or equal to 5')
    elif x<5:
        print('x is less than 5')
elif x>11:
    if x>50:
        print('x is greater than 50')
    elif x<=50:
        print('x is greater than 11 but less than or equal to 50')
```
