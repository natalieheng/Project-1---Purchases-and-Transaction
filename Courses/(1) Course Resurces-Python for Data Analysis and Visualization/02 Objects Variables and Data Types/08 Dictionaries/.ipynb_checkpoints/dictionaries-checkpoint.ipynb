## Dictionaries

Dictionaries store data in key value pairs, these key value pairs are written as key:value (values of dictionary can be any data type) 

Dictionaries are ordered (as of Python 3.7), mutable and do not allow duplicate objects

```python
dict_1 = {'brand':'Ford','cost':25000}
```

```python
dict_1
```

```python
# You can return the associated value for a specific key
dict_1['brand']
```

```python
# You can also update a value
dict_1['brand'] = 'Audi'
```

```python
dict_1
```

```python
# You can add a new key value pair as follows
dict_1['transmission']='Manual'
```

```python
dict_1
```

```python
# The value can be an embedded dictionary object
dict_1['other_attributes']={'colour':'Black','year':2021}
```

```python
dict_1
```

```python
dict_1['other_attributes']['year']
```

```python
# The value can also be a list object
dict_1['previous_owners']=['Mike','Shelly','Nitin']
```

```python
dict_1
```

```python
dict_1['previous_owners']
```

```python
x = dict_1['previous_owners']
```

```python
x[1]
```

```python
# You can stack index operations to return values from embedded list/dict objects
dict_1['previous_owners'][1]
```

## Dictionary Methods

```python
# The pop method must reference a key
dict_1.pop('previous_owners')
```

```python
dict_1
```

```python
# The items method will display key:value pairs as tuples
dict_1.items()
```

```python
# The keys method will return a list of all the keys
dict_1.keys()
```

```python
# The values method will return a list of all values
dict_1.values()
```
