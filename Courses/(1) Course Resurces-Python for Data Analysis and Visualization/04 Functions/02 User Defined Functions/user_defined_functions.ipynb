# User Defined Functions

```python
# Defining a function that prints 'welcome' when called
def welcome_message():
    print('welcome')
```

```python
welcome_message()
```

```python
# The function prints a message using an f string and refers to the value stored in x
def welcome_message(x):
    print(f'welcome {x}')
```

```python
welcome_message('John')
```

```python
# One argument is required unless you specify a default value for this argument on creation of the function
welcome_message()
```

```python
# Defining a function where 3 arguments are required
def welcome_message(x,y,z):
    print(f'welcome {x}, {y} and {z}')
```

```python
welcome_message('John','Anisha','Nitin')
```

```python
# Setting a default value for the arguments
def welcome_message(x = 'name1', y = 'name2', z = 'name3'):
    print(f'welcome {x}, {y} and {z}')
```

```python
welcome_message(z='John')
```

## Print vs Return

```python
def print_message():
    print('hello this message has been printed')
```

```python
def return_message():
    return 'hello this message has been returned'
```

```python
print_message()
```

```python
return_message()
```

```python
x= print_message()
```

```python
x
```

```python
# printing does not output anything
type(x)
```

```python
# ... But return does
x = return_message()
```
