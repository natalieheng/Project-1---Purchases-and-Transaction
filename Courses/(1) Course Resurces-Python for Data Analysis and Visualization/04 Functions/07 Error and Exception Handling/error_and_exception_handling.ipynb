# try except else finally
• The try block tries to execute a block of code

• The except block runs a block of code if an error is raised

• You can also provide an else statement

• The finally block runs regardless of an error

```python
print(100/10)
```

```python
# This raises a ZeroDivisionError exception
100/0
```

```python
# If you add a try block on its own you will still get an error
try:
    print(100/0)
```

```python
# If you add an except block a block will run if the try block raises an exception
try:
    print(100/0)
except:
    print('Something went wrong')
```

```python
print(100/0)
```

```python
# The except block is specifically handling ZeroDivisionError exceptions now
try:
    print(100/0)
except ZeroDivisionError:
    print('You cannot divide by zero')
```

```python
# The else block runs only if the try block succeeds
try:
    print(100/3)
except ZeroDivisionError:
    print('You cannot divide by zero')
else:
    print('Everything went well')
```

```python
# The finally block always executes
try:
    print(100/'hello')
except ZeroDivisionError:
    print('You cannot divide by zero')
else:
    print('Everything went well')
finally:
    print('this always prints')
```

```python
try:
    print(100/'hello')
except ZeroDivisionError:
    print('You cannot divide by zero')
else:
    print('Everything went well')
finally:
    print('this always prints')
```

```python
# You can add multiple except blocks to handle different exceptions
try:
    print(100/'hello')
except ZeroDivisionError:
    print('You cannot divide by zero')
except TypeError:
    print('incompatable types')
else:
    print('Everything went well')
finally:
    print('this always prints')
```
