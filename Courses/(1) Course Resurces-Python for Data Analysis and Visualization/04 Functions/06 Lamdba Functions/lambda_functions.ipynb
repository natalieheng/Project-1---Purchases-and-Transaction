# Lambda Expressions

A Lambda Function is an anonomous function that can take in any number of arguments and returns the output of an expression.

It works just like any other function except that you don't have to define a function name. The benefit of lambda expressions is in its simplicity, it only requires one line of code and the syntax is like this...

### Removing strings from a list without lambdas

```python
list_1 = ['Argentina', 'Mexico','Brazil',99]
```

```python
def is_str(x):
    return type(x)==str
```

```python
list(filter(is_str,list_1))
```

### Removing strings from a list with lambdas

```python
lambda x: type(x)==str
```

```python
list(filter(lambda x: type(x)==str,list_1))
```

### Multiplying elements in a list by 2 without lambdas

```python
list_2 = [1,2,3,4,5,6]
```

```python
def func_multiply(x):
    return x*2
```

```python
list(map(func_multiply,list_2))
```

### Multiplying elements in a list by 2 with lambdas

```python
list(map(lambda x: x*2,list_2))
```

### You cannot use lamdbas for scenarios that require multiple expressions

```python
def if_multiply(x):
    if x>3:
        return x*2
    else:
        return x
```

```python
list(map(if_multiply,list_2))
```
