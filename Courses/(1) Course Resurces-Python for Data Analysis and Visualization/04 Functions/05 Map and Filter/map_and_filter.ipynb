# Map and Filter

The map function can take an iterable element (such as a list) and it can apply a function to each element of that iterable object

It takes two arguments, the function to apply to each element, and the iterable object. 

The map function returns a map object which can be converted back to that iterable object or even looped to return each element with the function appled to it. 

```python
countries_list = ['Argentina','United States','Uganda','Germany']
```

```python
for x in countries_list:
    print(len(x))
```

```python
map(len,countries_list)
```

```python
list(map(len,countries_list))
```

```python
Argentina: 9
```

```python
def return_len(x):
    return f'{x}: {len(x)}'
```

```python
return_len('Hello')
```

```python
list(map(return_len,countries_list))
```

The filter function takes the same arguments as the map function but it works differently. It takes a function that returns a boolean value of True or False for each element of the iterable object. 

If the function returns True for that element then the element is returned, otherwise it is dropped from the output

```python
list_1 = ['Argentina','Mexico','Brazil',2]
```

```python
def is_str(x):
    return type(x)==str
```

```python
is_str(2)
```

```python
is_str('Argentina')
```

```python
filter(is_str,list_1)
```

```python
list(filter(is_str,list_1))
```

```python
list(map(is_str,list_1))
```
